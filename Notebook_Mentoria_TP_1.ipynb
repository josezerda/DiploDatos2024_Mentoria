**Cargamos temperaturas maximas, medias y minimas**

```python
df_temp_max = pd.read_csv('temp_max.txt', sep=';')
df_temp_media = pd.read_csv('temp_media.txt', sep=';')
df_temp_min = pd.read_csv('temp_min.txt', sep=';')
```

Podemos crear un dataframe en donde una columa sea la fecha del año y el otra columna la temperatura

```python
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
```

```python
temperaturas_maximas = []
temperaturas_medias = []
temperaturas_minimas = []
```

**Colocamos las temperaturas en una lista desde el 1/1/20233 al 31/12/2023**

```python
for valor in meses:
    temp_mes = df_temp_max[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_maximas.extend(lista_aux) 
    #print(valor)
for x in temperaturas_maximas[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_maximas.remove(x)  # Eliminamos los datos 's/d'
```

```python
for valor in meses:
    temp_mes = df_temp_media[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_medias.extend(lista_aux) 
    #print(valor)
for x in temperaturas_medias[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_medias.remove(x)  # Eliminamos los datos 's/d'
```

```python
for valor in meses:
    temp_mes = df_temp_min[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_minimas.extend(lista_aux) 
    #print(valor)
for x in temperaturas_minimas[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_minimas.remove(x)  # Eliminamos los datos 's/d'
```

```python
fechas_2023 = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
df_fechas_temperaturas = pd.DataFrame({'Fecha': fechas_2023, 'Temp_max': temperaturas_maximas, 'Temp_media': temperaturas_medias, 'Temp_min': temperaturas_minimas})
```

```python
df_fechas_temperaturas
```

**Paso los valores de temperatura a numericos**

```python
df_fechas_temperaturas['Temp_max'] = df_fechas_temperaturas['Temp_max'].str.replace(',', '.').astype(float)
df_fechas_temperaturas['Temp_media'] = df_fechas_temperaturas['Temp_media'].str.replace(',', '.').astype(float)
df_fechas_temperaturas['Temp_min'] = df_fechas_temperaturas['Temp_min'].str.replace(',', '.').astype(float)
```

```python
df_melted = pd.melt(df_fechas_temperaturas, id_vars=['Fecha'], value_vars=['Temp_max', 'Temp_media', 'Temp_min'], var_name='Tipo', value_name='Temperatura')

plt.figure(figsize=(20, 6))
sns.lineplot(data=df_melted, x='Fecha', y='Temperatura', hue='Tipo')

plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.title('Temperaturas en el tiempo')
plt.grid(True)
plt.show()
```

Vamos a recorrer el dataset a fin de calcular la energia diaria para cada una de las energias catalogadas

```python
columnas_deseadas = ['ENERGIA_DIRECTA', 'ENERGIA_RESTO', 'ENERGIA_PICO', 'ENERGIA_VALLE']
nombres_columnas_diferencias = ['DIF_ENERGIA_DIRECTA', 'DIF_ENERGIA_RESTO', 'DIF_ENERGIA_PICO', 'DIF_ENERGIA_VALLE']
lista_dfs_diferencias = []

cant_dias_mediciones = list(range(330, 366))

for valor in cant_dias_mediciones:
    conteo_valores = df_diario_combinado_renombrado['MEDIDOR'].value_counts()
    valores_repetidos = conteo_valores[conteo_valores == valor].index.tolist()

    for medidor in valores_repetidos:
        df_medidor = df_diario_combinado_renombrado[df_diario_combinado_renombrado['MEDIDOR'] == medidor]

        # Calcular diferencias y crear un DataFrame pequeño
        diferencias_medidor = df_medidor[columnas_deseadas].iloc[1:].subtract(df_medidor[columnas_deseadas].shift().iloc[1:])
        diferencias_medidor.columns = nombres_columnas_diferencias  # Renombrar columnas
        diferencias_medidor['LECTURA'] = df_medidor['LECTURA'].iloc[1:]
        diferencias_medidor['MEDIDOR'] = medidor

        # Agregar el DataFrame pequeño a la lista
        lista_dfs_diferencias.append(diferencias_medidor)

# Concatenar todos los DataFrames pequeños en uno grande
diferencias_total = pd.concat(lista_dfs_diferencias, ignore_index=True)

# Unir las diferencias al DataFrame principal
df_diario_combinado_renombrado = df_diario_combinado_renombrado.merge(
    diferencias_total, on=['LECTURA', 'MEDIDOR'], how='left'
)   
    
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que df_diario_combinado_renombrado ya está definido y contiene las columnas 'LECTURA' y 'DIF_ENERGIA_DIRECTA'

# Asegurarse de que la columna 'LECTURA' sea de tipo datetime
df_diario_combinado_renombrado['LECTURA'] = pd.to_datetime(df_diario_combinado_renombrado['LECTURA'])

# Crear columnas adicionales para el heatmap, por ejemplo, mes y día
df_diario_combinado_renombrado['Mes'] = df_diario_combinado_renombrado['LECTURA'].dt.month
df_diario_combinado_renombrado['Día'] = df_diario_combinado_renombrado['LECTURA'].dt.day

# Crear una tabla pivote
pivot_table = df_diario_combinado_renombrado.pivot_table(
    values='DIF_ENERGIA_DIRECTA',
    index='Mes',
    columns='Día',
    aggfunc=np.sum
)

# Graficar el heatmap
plt.figure(figsize=(30, 10))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".1f")
plt.title('Mapa de Calor de DIF_ENERGIA_DIRECTA a lo largo del Año')
plt.xlabel('Día del Mes')
plt.ylabel('Mes')
plt.show()
```
