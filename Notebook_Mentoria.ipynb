## **Al dataframe combinado le vamos a sumar las siguientes columnas:**

1. Temperatura maxima del dia
2. Temperatura media del dia
3. Temperatura minima del dia

**Cargamos temperaturas maximas, medias y minimas**

```python
df_temp_max = pd.read_csv('temp_max.txt', sep=';')
df_temp_media = pd.read_csv('temp_media.txt', sep=';')
df_temp_min = pd.read_csv('temp_min.txt', sep=';')
```

Podemos crear un dataframe en donde una columa sea la fecha del año y el otra columna la temperatura

```python
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
```

```python
temperaturas_maximas = []
temperaturas_medias = []
temperaturas_minimas = []
```

**Colocamos las temperaturas en una lista desde el 1/1/20233 al 31/12/2023**

```python
for valor in meses:
    temp_mes = df_temp_max[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_maximas.extend(lista_aux) 
    #print(valor)
for x in temperaturas_maximas[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_maximas.remove(x)  # Eliminamos los datos 's/d'
```

```python
for valor in meses:
    temp_mes = df_temp_media[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_medias.extend(lista_aux) 
    #print(valor)
for x in temperaturas_medias[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_medias.remove(x)  # Eliminamos los datos 's/d'
```

```python
for valor in meses:
    temp_mes = df_temp_min[valor]
    lista_aux = temp_mes.tolist()
    temperaturas_minimas.extend(lista_aux) 
    #print(valor)
for x in temperaturas_minimas[:]:  # Iteramos sobre la lista
    if x == 's/d':
        temperaturas_minimas.remove(x)  # Eliminamos los datos 's/d'
```

```python
fechas_2023 = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
df_fechas_temperaturas = pd.DataFrame({'Fecha': fechas_2023, 'Temp_max': temperaturas_maximas, 'Temp_media': temperaturas_medias, 'Temp_min': temperaturas_minimas})
```

```python
df_fechas_temperaturas
```

**Paso los valores de temperatura a numericos**

```python
df_fechas_temperaturas['Temp_max'] = df_fechas_temperaturas['Temp_max'].str.replace(',', '.').astype(float)
df_fechas_temperaturas['Temp_media'] = df_fechas_temperaturas['Temp_media'].str.replace(',', '.').astype(float)
df_fechas_temperaturas['Temp_min'] = df_fechas_temperaturas['Temp_min'].str.replace(',', '.').astype(float)
```

```python
df_melted = pd.melt(df_fechas_temperaturas, id_vars=['Fecha'], value_vars=['Temp_max', 'Temp_media', 'Temp_min'], var_name='Tipo', value_name='Temperatura')

plt.figure(figsize=(20, 6))
sns.lineplot(data=df_melted, x='Fecha', y='Temperatura', hue='Tipo')

plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.title('Temperaturas en el tiempo')
plt.grid(True)
plt.show()
```
